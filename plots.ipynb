```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import os
import shutil
import matplotlib

from copy import deepcopy

from plotting_functions import plot_experiment, plot_moving_compare

import matplotlib.pyplot as plt

def plot_aunlc_barchart(data, save_path, colors=None, labels=None):
    """
    Plots a bar chart with the given data, colors, and labels.

    Parameters:
    data (dict): Dictionary with keys as labels and values as heights of the bars
    colors (list, optional): List of colors for each bar
    labels (list, optional): List of labels for each bar
    """
    keys = list(data.keys())
    values = list(data.values())

    if labels is None:
        labels = keys
    if colors is None:
        colors = ['blue'] * len(keys)  # Default color is blue

    plt.figure(figsize=(6, 6))

    # Creating the modified positions with space after every third bar
    positions = []
    for i in range(len(keys)):
        positions.append(i + (i // 2) * 1.0)  # Add a space after every third bar


    bars = plt.bar(positions, values, color=colors, tick_label=labels)
    plt.xlabel(None)
    plt.ylabel('AUNLC', fontsize=12)
    plt.title('Area under normalized learning curves', fontsize=12)
    plt.yticks(fontsize=12)
    plt.xticks(positions, labels, rotation=0, ha='center', fontsize=12)

    # Adding value labels on top of the bars
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 3), va='bottom', ha='center', fontsize=12)

    plt.tight_layout()
    plt.savefig(save_path , dpi=96)


def unnormalize_array(array, min_val, max_val, k):
    """
    Rescale the array such that new_arr[0] = min_val and new_arr[k] = max_val.
    
    Parameters:
    array (numpy.ndarray): The input array to rescale.
    min_val (float): The desired value for the first element of the new array.
    max_val (float): The desired value for the k-th element of the new array.
    k (int): The index in the array which should correspond to max_val after rescaling.
    
    Returns:
    numpy.ndarray: The rescaled array.
    """
    if not 0 <= k < len(array):
        raise ValueError("k must be within the range of the array's indices.")
    
    # Determine the scale and offset for linear transformation
    old_start = array[0]
    old_end = array[k]
    scale = (max_val - min_val) / (old_end - old_start)
    offset = min_val - scale * old_start

    # Apply the linear transformation
    new_array = scale * array + offset
    
    return new_array


def rescale_array(x, K):
    """
    Rescale a numpy array x such that x'[0] = 0 and x'[K] = 1.

    Parameters:
    x (numpy array): The input array.
    K (int): The index to be scaled to 1.

    Returns:
    numpy array: The rescaled array.
    """
    # Check if K is within the bounds of the array
    if K >= len(x):
        raise ValueError("K must be within the bounds of the array length.")

    # Subtract the first element from all elements
    x_rescaled = x - x[0]

    # Divide all elements by the K-th element
    x_rescaled = x_rescaled / x_rescaled[K]

    return x_rescaled
    
custom_colors = {
    'darkblue': '#0415cc',
    'black': '#000000',
    'lightblue': '#48cefa',
    'blue': '#0275f7',
    'orange': '#FFAA00',
    'green': '#11CC11',
    'darkgreen': "#016103",
    'pink': '#f702df',
    'purple': '#8902f7',
    'red': '#CC0022',
    'yellow': '#F0D919',
}

figpath = './final_figs_combined'

try:
    shutil.rmtree(figpath)
    os.makedirs(figpath)
except:
    os.makedirs(figpath)
```

```python
def compute_auc(x_vec,y_vec):
    return np.trapz(y_vec, x_vec)
    
def get_k_threshold(y_vec, threshold):
    """
    Find the lowest index k such that y_vec[k] >= threshold.

    Parameters:
    y_vec (list or array): The list or array of y values.
    threshold (numeric): The threshold value to compare against.

    Returns:
    int: The lowest index k such that y_vec[k] >= threshold. Returns -1 if no such index exists.
    """
    for k, value in enumerate(y_vec):
        if value >= threshold:
            return k
    return -1  # Return -1 if no such index exists

qubit_thresholds = dict()
for key in qubit_plotted_data.keys():
    k = get_k_threshold(qubit_plotted_data[key]["y"], 0.90)
    qubit_thresholds[key] = k
    print("QUBIT:", key, "=>", k)

piquasso_thresholds = dict()
for key in piquasso_plotted_data.keys():
    k = get_k_threshold(piquasso_plotted_data[key]["y"], 0.90)
    piquasso_thresholds[key] = k
    print("PIQUASSO:", key, "=>", k)

# Smallest value larger than -1.
qubit_threshold_cp = sorted([qubit_thresholds[key] for key in ['CP, joint', 'CP, fixed'] if qubit_thresholds[key] > -1])[0]
qubit_threshold_mz = sorted([qubit_thresholds[key] for key in ['MZ, joint', 'MZ, fixed'] if qubit_thresholds[key] > -1])[0]

piquasso_threshold_cp = sorted([piquasso_thresholds[key] for key in ['CP, joint', 'CP, fixed'] if piquasso_thresholds[key] > -1])[0]
piquasso_threshold_mz = sorted([piquasso_thresholds[key] for key in ['MZ, joint', 'MZ, fixed'] if piquasso_thresholds[key] > -1])[0]

#TODO: Re-scale qubit plotted data and piquasso plotted data along x axis based on the newly computed thresholds.

print("qubit_threshold_cp =", qubit_threshold_cp)
print("qubit_threshold_mz =", qubit_threshold_mz)

print("piquasso_threshold_cp =", piquasso_threshold_cp)
print("piquasso_threshold_mz =", piquasso_threshold_mz)


print("--------------------")
print("qubit_plotted_data['CP, joint']['x'][qubit_threshold_cp] = ", qubit_plotted_data['CP, joint']['x'][qubit_threshold_cp])
print("qubit_plotted_data['CP, fixed']['x'][qubit_threshold_cp] = ", qubit_plotted_data['CP, fixed']['x'][qubit_threshold_cp])
print("qubit_plotted_data['MZ, joint']['x'][qubit_threshold_mz] = ", qubit_plotted_data['MZ, joint']['x'][qubit_threshold_mz])
print("qubit_plotted_data['MZ, fixed']['x'][qubit_threshold_mz] = ", qubit_plotted_data['MZ, fixed']['x'][qubit_threshold_mz])

print("--------------------")
print("piquasso_plotted_data['CP, joint']['x'][piquasso_threshold_cp] = ", piquasso_plotted_data['CP, joint']['x'][piquasso_threshold_cp])
print("piquasso_plotted_data['CP, fixed']['x'][piquasso_threshold_cp] = ", piquasso_plotted_data['CP, fixed']['x'][piquasso_threshold_cp])
print("piquasso_plotted_data['MZ, joint']['x'][piquasso_threshold_mz] = ", piquasso_plotted_data['MZ, joint']['x'][piquasso_threshold_mz])
print("piquasso_plotted_data['MZ, fixed']['x'][piquasso_threshold_mz] = ", piquasso_plotted_data['MZ, fixed']['x'][piquasso_threshold_mz])


print("--------------------")
print("qubit_plotted_data['CP, joint']['x'][-1] = ", qubit_plotted_data['CP, joint']['x'][-1])
print("qubit_plotted_data['CP, fixed']['x'][-1] = ", qubit_plotted_data['CP, fixed']['x'][-1])
print("qubit_plotted_data['MZ, joint']['x'][-1] = ", qubit_plotted_data['MZ, joint']['x'][-1])
print("qubit_plotted_data['MZ, fixed']['x'][-1] = ", qubit_plotted_data['MZ, fixed']['x'][-1])

print("--------------------")
print("piquasso_plotted_data['CP, joint']['x'][-1] = ", piquasso_plotted_data['CP, joint']['x'][-1])
print("piquasso_plotted_data['CP, fixed']['x'][-1] = ", piquasso_plotted_data['CP, fixed']['x'][-1])
print("piquasso_plotted_data['MZ, joint']['x'][-1] = ", piquasso_plotted_data['MZ, joint']['x'][-1])
print("piquasso_plotted_data['MZ, fixed']['x'][-1] = ", piquasso_plotted_data['MZ, fixed']['x'][-1])


rescaled_qubit_plotted_data = deepcopy(qubit_plotted_data)

rescaled_piquasso_plotted_data = deepcopy(piquasso_plotted_data)

rescaled_qubit_plotted_data['CP, joint']['x'] = rescale_array(np.array(qubit_plotted_data['CP, joint']['x']), qubit_threshold_cp)
rescaled_qubit_plotted_data['CP, fixed']['x'] = rescale_array(np.array(qubit_plotted_data['CP, fixed']['x']), qubit_threshold_cp)
rescaled_qubit_plotted_data['MZ, joint']['x'] = rescale_array(np.array(qubit_plotted_data['MZ, joint']['x']), qubit_threshold_mz)
rescaled_qubit_plotted_data['MZ, fixed']['x'] = rescale_array(np.array(qubit_plotted_data['MZ, fixed']['x']), qubit_threshold_mz)

rescaled_piquasso_plotted_data['CP, joint']['x'] = rescale_array(np.array(piquasso_plotted_data['CP, joint']['x']), piquasso_threshold_cp)
rescaled_piquasso_plotted_data['CP, fixed']['x'] = rescale_array(np.array(piquasso_plotted_data['CP, fixed']['x']), piquasso_threshold_cp)
rescaled_piquasso_plotted_data['MZ, joint']['x'] = rescale_array(np.array(piquasso_plotted_data['MZ, joint']['x']), piquasso_threshold_mz)
rescaled_piquasso_plotted_data['MZ, fixed']['x'] = rescale_array(np.array(piquasso_plotted_data['MZ, fixed']['x']), piquasso_threshold_mz)

qubit_plotted_data['CP, joint']['x'] = np.array(qubit_plotted_data['CP, joint']['x'])
qubit_plotted_data['CP, fixed']['x'] = np.array(qubit_plotted_data['CP, fixed']['x'])
qubit_plotted_data['MZ, joint']['x'] = np.array(qubit_plotted_data['MZ, joint']['x'])
qubit_plotted_data['MZ, fixed']['x'] = np.array(qubit_plotted_data['MZ, fixed']['x'])

piquasso_plotted_data['CP, joint']['x'] = np.array(piquasso_plotted_data['CP, joint']['x'])
piquasso_plotted_data['CP, fixed']['x'] = np.array(piquasso_plotted_data['CP, fixed']['x'])
piquasso_plotted_data['MZ, joint']['x'] = np.array(piquasso_plotted_data['MZ, joint']['x'])
piquasso_plotted_data['MZ, fixed']['x'] = np.array(piquasso_plotted_data['MZ, fixed']['x'])
```
